```python
from collections import defaultdict
def group_words_by_line(words, tolerance=5):
    """Group words by their vertical position with tolerance"""
    lines = defaultdict(list)
    for word in words:
        # Use rounded vertical position as key
        key = round(word['top'] / tolerance) * tolerance
        lines[key].append(word)
    return lines
```
